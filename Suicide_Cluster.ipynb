# **Method**
 

*   To quantify the data correlation between two files,the **Pearson Correlation** method is used.  Pearson Correlation method is a methor to measure of the strength of a linear association between two variables and is denoted by r which indicates how far away all these data points are to this line of best fit (i.e., how well the data points fit this new model/line of best fit). Furthermore, **Granger Causality test** is also be used to check the causality effect between data. Granger causality is a statistical concept of causality that is based on prediction. According to Granger causality, if a signal X1 "Granger-causes" (or "G-causes") a signal X2, then past values of X1 should contain information that helps predict X2 above and beyond the information contained in past values of X2 alone. Its mathematical formulation is based on linear regression modeling of stochastic processes (Granger 1969)

* To visualise the countries pattern, this project use  **Geopandas** mapping library. 

*   For clustering the countries, **Clustering Algorithm** is chosen as it is unsupervised learning which figures out the pattern of the subjects so that will be able to group different subjects into different clusters. The specific algorithhm is the **KMeans clustering**.  KMeans clustering is  a method of vector quantization, originally from signal processing, that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster. This results in a partitioning of the data space into Voronoi cells.

To ensure the relationship between both data, causality test is also executed below.

```python
#Granger Causality Test
import pandas as pd
import statsmodels.tsa.stattools as sts

# Load the data
df = df_agg

# Perform the Granger causality test
result= sts.grangercausalitytests(df[['InternetPercentage', 'SuicideRate']], maxlag=2)

# Print the test statistic and p-value
print(result[2][0])
```

The F test statistic turns out to be **0.48** and the corresponding p-value is **0.65**

Since the p-value is more than .05, we can not reject the null hypothesis of the test and conclude that **there is a causal relationship** between both data and knowing the number of internet user percentage is useful for predicting the future suicide rate.

**Thus, the overall conclusion is, although the high internet user rate in a country does not directly mean it will have a high suicide number, it is still a playing factor that cause a big suicide number.**

# **Modelling**

Due to the different behavior on both data among countries, in this part we will cluster the countries based on both data to identify any communal pattern among them and figure out how exactly their likelihood are.

This part will use KMeans as it is the most used algorithm to cluster data.

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import davies_bouldin_score
```

```python
#Normalize data into 0-1 using Percentile Linearization Max
data = df_agg_countries
data.index = df_agg_countries['country']
size = len(data)-1
input = data[['InternetPercentage','SuicideRate']].rank(method='max').apply(lambda x: (x-1)/size)
input.describe()
```

```python
input.head()
```

```python
#Get Davied Bouldin Score to get the best cluster number
cluster = {}
for k in range(2, 10):
    print("K = {}".format(k))
    kmeans = KMeans(n_clusters=k, max_iter=1000, random_state=42).fit(input)
    label = kmeans.labels_
    
    dbi = davies_bouldin_score(input, label)
    cluster[k] = {
        "inertia": kmeans.inertia_, # Inertia: Sum of distances of samples to their closest cluster center,
        "label": label,
        "dbi": dbi
    }
    print("For n_clusters={}, The davies bouldin score is {}. The inertia is {}".format(k, dbi, cluster[k]['inertia']))
```

```python
inertia = []
for k,v in cluster.items():
    inertia.append(v['inertia'])
    
plt.figure(figsize=(8,5))
plt.plot(list(cluster.keys()), list(inertia), marker='o')
plt.xlabel("Number of cluster")
plt.ylabel("Inertia")
plt.show()
```

To this step, we are using the elbow method to pre-determine the best cluster number. Elbow method is a heuristic used in determining the number of clusters in a data set by picking the elbow of the curve as the number of clusters to use.

Thus, for this data, we see the curve elbow is amongst number 3, 4 and 5

```python
dbi = []
for k,v in cluster.items():
    dbi.append(v['dbi'])
    
plt.figure(figsize=(8,5))
plt.plot(list(cluster.keys()), list(dbi), marker='o')
plt.xlabel("Number of cluster")
plt.ylabel("DBI")
plt.show()
```

The metric is Davies Bouldin that is defined as the average similarity measure of each cluster with its most similar cluster, where similarity is the ratio of within-cluster distances to between-cluster distances. The minimum score is zero, with lower values indicating better clustering.

By using elbow result (3,4,5), we see that number 4 has the smaller DBI score. Therefore for this model, **4 clusters is the best**.

```python
#Run model
kmeans = KMeans(n_clusters=4, random_state=0).fit(input)
idx = np.argsort(kmeans.cluster_centers_.sum(axis=1))
lut = np.zeros_like(idx)
lut[idx] = np.arange(4)
labels = lut[kmeans.labels_]
input['cluster'] = labels
input = input.sort_values(['cluster'])
input
```
