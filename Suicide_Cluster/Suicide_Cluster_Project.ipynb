## **2.Social needs and social risks**

Social communication is human nature, not only the utilitarian needs, more important is the psychological needs. People get a sense of security and satisfaction in social activities, vent emotional pressure and so on. This can all be done more conveniently on social media platforms. # **Model**
 

*   We want to divide the countries into two groups, the ones with high number of suicides and the ones with low number of suicides, to show the correlation between suicide number and countries.
*   So we intent to choose **Clustering Algorithm**. Cluster Algorithm is unsupervised learning which figures out the pattern of the subjects so that will be able to group different subjects into different clusters.

*   We plan to choose **K-means clustering**. 

*   **k-means clustering** is a method of vector quantization, that aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster. This results in a partitioning of the data space into Voronoi cells. 

*   **K-means clustering**  was easy to implement in this case, since the number of suicides committed by people in different countries, so we didn't have to pre-specify the number of clusters(k).
